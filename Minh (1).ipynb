```
demo_merge_6 = pd.read_csv('gdrive/My Drive/Datathon Citizens/combined_all_month.csv')
demo_merge_6.head()
```

```
col_list = list(demo_merge_6.columns)
pd.DataFrame(col_list)
```

```
missing_value = demo_merge_6.isnull().sum()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
mv_perc = missing_value/len(demo_merge_6)
list_of_tuples = list(zip(col_list,missing_value, mv_perc)) 
demo_merge_6_mv_df = pd.DataFrame(data=list_of_tuples, columns = ['Columns','Missing values', 'Percentage'])
demo_merge_6_mv_df
```

```
rslt_df = demo_merge_6_mv_df[demo_merge_6_mv_df['Percentage'] <= 0.05]
mv_less_than_5 = list(rslt_df['Columns'])
mv_less_than_5.remove('Unnamed: 0')
mv_less_than_5
```

#### get rid of all zip columns because we don't want to impute them

```
mv_less_than_5 = [c for c in mv_less_than_5 if 'zip' not in c]
mv_less_than_5
```

```
for i in mv_less_than_5: 
    mean = demo_merge_6[i].mean()
    demo_merge_6[i] = demo_merge_6[i].fillna(mean)

demo_merge_6.head()
```

```
missing_value = demo_merge_6.isnull().sum()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
mv_perc = missing_value/len(demo_merge_6)
list_of_tuples = list(zip(col_list,missing_value, mv_perc)) 
demo_merge_6_mv_df = pd.DataFrame(data=list_of_tuples, columns = ['Columns','Missing values', 'Percentage'])
demo_merge_6_mv_df
```

##### **Now just fill out all NA of morgages columns with 0**

```
mortgage_columns_1 = demo_merge_6_mv_df[demo_merge_6_mv_df.Columns.str.contains('mortgage',case=False)]
mortgage_columns_1 = mortgage_columns_1[mortgage_columns_1['Percentage'] > 0]
mortgage_columns_1.shape
```

```
mortgage_columns = list(mortgage_columns_1['Columns'])
len(mortgage_columns)
```

```
for i in mortgage_columns: 
    demo_merge_6[i] = demo_merge_6[i].fillna(0)

demo_merge_6.head()
```

```
demo_merge_6.shape
```

```
missing_value = demo_merge_6.isnull().sum()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
mv_perc = missing_value/len(demo_merge_6)
list_of_tuples = list(zip(col_list,missing_value, mv_perc)) 
demo_merge_6_mv_df = pd.DataFrame(data=list_of_tuples, columns = ['Columns','Missing values', 'Percentage'])
demo_merge_6_mv_df
```

```
homeequity_columns_3 = demo_merge_6_mv_df[demo_merge_6_mv_df.Columns.str.contains('homeequity3',case=False)]
homeequity_columns_4 = demo_merge_6_mv_df[demo_merge_6_mv_df.Columns.str.contains('homeequity4',case=False)]
homeequity_columns_5 = demo_merge_6_mv_df[demo_merge_6_mv_df.Columns.str.contains('homeequity5',case=False)]

homeequity_columns_3 = homeequity_columns_3[homeequity_columns_3['Percentage'] > 0]
homeequity_columns_4 = homeequity_columns_4[homeequity_columns_4['Percentage'] > 0]
homeequity_columns_5 = homeequity_columns_5[homeequity_columns_5['Percentage'] > 0]

homeequity_columns_3.shape
```

```
homeequity_columns_3 = list(homeequity_columns_3['Columns'])
homeequity_columns_4 = list(homeequity_columns_4['Columns'])
homeequity_columns_5 = list(homeequity_columns_5['Columns'])

len(homeequity_columns_3)
```

```
for i in homeequity_columns_3: 
    demo_merge_6[i] = demo_merge_6[i].fillna(0)

for i in homeequity_columns_4: 
    demo_merge_6[i] = demo_merge_6[i].fillna(0)

for i in homeequity_columns_5: 
    demo_merge_6[i] = demo_merge_6[i].fillna(0)

demo_merge_6.head()
```

```
missing_value = demo_merge_6.isnull().sum()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
mv_perc = missing_value/len(demo_merge_6)
list_of_tuples = list(zip(col_list,missing_value, mv_perc)) 
demo_merge_6_mv_df = pd.DataFrame(data=list_of_tuples, columns = ['Columns','Missing values', 'Percentage'])
demo_merge_6_mv_df
```

```
giving_up = demo_merge_6_mv_df[demo_merge_6_mv_df['Percentage'] > 0]
giving_up = list(giving_up['Columns'])
len(giving_up)
```

```
demo_merge_6 = demo_merge_6.drop(columns=giving_up)
demo_merge_6.shape
```

```
y_train = demo_merge_6['homebuyers']
X_train = demo_merge_6.drop('homebuyers', axis=1)
X_train.shape
```

```
from sklearn.ensemble import RandomForestClassifier
 

rnd_clf = RandomForestClassifier(n_estimators=5, n_jobs=-1)
rnd_clf.fit(X_train, y_train)
for name, score in zip(demo_merge_6["feature_names"], rnd_clf.feature_importances_):
    print(name, score)
```
