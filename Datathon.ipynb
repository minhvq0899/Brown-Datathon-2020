```python
#may_df = pd.read_csv(root_path + "zip9_coded_201905_pv.csv")
demo_df = pd.read_csv(root_path + "zip9_demographics_coded_pv.csv")

demo_df['homebuyers'][demo_df['homebuyers'] >= 2] = 2
split = StratifiedShuffleSplit(n_splits=1, test_size=0.01, random_state=42)
for train_index, test_index in split.split(demo_df, demo_df["homebuyers"]):
    demo_left_out = demo_df.loc[train_index]
    demo_random_subset = demo_df.loc[test_index]

shape = (demo_left_out.shape, demo_random_subset.shape)
shape
```

```python
apr_df = pd.read_csv(root_path + "zip9_coded_201904_pv.csv")
may_df = pd.read_csv(root_path + "zip9_coded_201905_pv.csv")
jun_df = pd.read_csv(root_path + "zip9_coded_201906_pv.csv")
jul_df = pd.read_csv(root_path + "zip9_coded_201907_pv.csv")
aug_df = pd.read_csv(root_path + "zip9_coded_201908_pv.csv")
sep_df = pd.read_csv(root_path + "zip9_coded_201909_pv.csv")
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
imputer = IterativeImputer(estimator=RandomForestRegressor(),random_state=1000)
column_names = demo_merge_6.columns
missing_value = demo_merge_6.isnull().sum()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
mv_perc = missing_value/len(demo_merge_6)
list_of_tuples = list(zip(column_names,missing_value, mv_perc)) 
train_mv_df = pd.DataFrame(data=list_of_tuples, columns = ['Columns','Missing values', 'Percentage'])
train_mv_df
```

```python
#missing_less_than_5 = [x['Columns'] for x in train_mv_df if float(x['Percentage']) < 0.05]

missing_less_than_5 = []
for i in range(len(train_mv_df)):
    if train_mv_df['Percentage'][i]<0.05:
        missing_less_than_5.append(train_mv_df['Columns'][i])
impute_columns = imputer.fit_transform(demo_merge_6[missing_less_than_5])
```
